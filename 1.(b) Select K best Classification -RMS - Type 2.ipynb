```python
# Loading the needed packages/lib
import pandas as pd
from sklearn.model_selection import train_test_split 
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.metrics import mean_squared_error
from sklearn.feature_selection import f_classif
from sklearn.linear_model import LogisticRegression
import pickle
import matplotlib.pyplot as plt


# call this func selectkbest and input x and y variable along with declaring no of variables
def selectkbest(indep_X,dep_Y,n): 
        #calling selectKbest algo + eval metrics = Chi square + decalring no of features
        test = SelectKBest(score_func=f_classif, k=n)
        #fit the model selectKbest & input indep, dep variable
        fit1= test.fit(indep_X,dep_Y) 
        #Transform selectKbest algo + eval metrics = Chisquare + declared no of features into indep_x variable
        selectk_features = fit1.transform(indep_X) 
        return selectk_features # finally return to selectk_features variable
    

#def custom_rms(y_test, y_pred):
    #return np.sqrt(mean_squared_error(y_test, y_pred))

    #best_features = []
    #best_rms = float('inf')  # Initialize with a high value

    #for feature_idx in range(X_train.shape[1]):
        # Select the current feature
        #selected_features = best_features + [feature_idx]
        #X_train_selected = X_train[:, selected_features]
        #X_test_selected = X_test[:, selected_features]
    
def split_scalar(indep_X,dep_Y):# call this func split_scalar and input x and y
        # Training & test split happening here
        X_train, X_test, y_train, y_test = train_test_split(indep_X, dep_Y, test_size = 0.25, random_state = 0)
        sc = StandardScaler()
        X_train = sc.fit_transform(X_train) # Fit & transform standardscalar ex from -1 to +1 or -2 or +2 to x_train
        X_test = sc.transform(X_test)# Alreday fitted so tranaform standardscalar ex from -1 to +1 or -2 or +2 to x_test
        return X_train, X_test, y_train, y_test # finally return to X_train, X_test, y_train, y_test.
    
def cm_prediction(classifier,X_test):# classifier is the variable holding model below(Log,DT,SVM,RF),x_test to predict y
     y_pred = classifier.predict(X_test) #Predicting Y variable using X_test

     from sklearn.metrics import confusion_matrix  # Import the Confusion Matrix from sklearn
     cm = confusion_matrix(y_test, y_pred) # creating confusion_matrix comparing Acutal vs predicted 
        
     from sklearn.metrics import accuracy_score  # Import the accuracy_score from sklearn
     from sklearn.metrics import classification_report #Import the classification_report from sklearn
        
     Accuracy=accuracy_score(y_test, y_pred )# creating accuracy_score comparing Acutal vs predicted 
        
     report=classification_report(y_test, y_pred) # creating Classification report comparing Acutal vs predicted 
    # finally return to classifier,Accuracy,report,X_test,y_test,cm so we can get scoring for each attributes  
     return  classifier,Accuracy,report,X_test,y_test,cm

def logistic(X_train,y_train,X_test):# Passing X_train,y_train to fit to the model and x_test to predict Y in cm_prediction
        # Fitting K-NN to the Training set
        from sklearn.linear_model import LogisticRegression
        classifier = LogisticRegression(random_state = 0)
        classifier.fit(X_train, y_train)
        # can get scoring either classifier or Accuracy or report or X_test or y_test or cm depending the way we call it finally
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm      
    
def svm_linear(X_train,y_train,X_test):# Passing X_train,y_train to fit to the model and x_test to predict Y in cm_prediction
                
        from sklearn.svm import SVC
        classifier = SVC(kernel = 'linear', random_state = 0)
        classifier.fit(X_train, y_train)
        # can get scoring classifier or Accuracy or report or X_test or y_test or cm or all depending the way we call it finally
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm
           # SAME APPLICABLE TO ALL OTHER ALGO BELOW
    
def svm_NL(X_train,y_train,X_test):
                
        from sklearn.svm import SVC
        classifier = SVC(kernel = 'rbf', random_state = 0)
        classifier.fit(X_train, y_train)
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm

def Navie(X_train,y_train,X_test):       
        # Fitting K-NN to the Training set
        from sklearn.naive_bayes import GaussianNB
        classifier = GaussianNB()
        classifier.fit(X_train, y_train)
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm         
    
    
def knn(X_train,y_train,X_test):
           
        # Fitting K-NN to the Training set
        from sklearn.neighbors import KNeighborsClassifier
        classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
        classifier.fit(X_train, y_train)
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm
def Decision(X_train,y_train,X_test):
        
        # Fitting K-NN to the Training set
        from sklearn.tree import DecisionTreeClassifier
        classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
        classifier.fit(X_train, y_train)
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm      


def random(X_train,y_train,X_test):
        
        # Fitting K-NN to the Training set
        from sklearn.ensemble import RandomForestClassifier
        classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
        classifier.fit(X_train, y_train)
        classifier,Accuracy,report,X_test,y_test,cm=cm_prediction(classifier,X_test)
        return  classifier,Accuracy,report,X_test,y_test,cm
    
def selectk_Classification(acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf): 
    # Create a Dataframe (table) with row ChiSquare and cols specified in cols
    dataframe=pd.DataFrame(index=['ChiSquare'],columns=['Logistic','SVMl','SVMnl','KNN','Navie','Decision','Random'])
    # Using For loop index placing the scoring from 0 to 6 to its respective algo
    for number,index in enumerate(dataframe.index):      
        dataframe['Logistic'][index]=acclog[number]       
        dataframe['SVMl'][index]=accsvml[number]
        dataframe['SVMnl'][index]=accsvmnl[number]
        dataframe['KNN'][index]=accknn[number]
        dataframe['Navie'][index]=accnav[number]
        dataframe['Decision'][index]=accdes[number]
        dataframe['Random'][index]=accrf[number]
    return dataframe # return the function to dataframe
```

```python
dataset1=pd.read_csv("prep.csv",index_col=None) # reading.loading the csv dataset 

df2=dataset1 # taking a back up of original dataset

df2 = pd.get_dummies(df2, drop_first=True) # remove first redundant col wrt to nominal col

indep_X=df2.drop('classification_yes', axis=1) # dropping 'classification_yes' from indep_x variable
dep_Y=df2['classification_yes'] # calling dep_y 'classification_yes'
```

```python
# calling SelectKBest func inputting indep_X,dep_Y and no of feature in args from sklearn.feature_selection import SelectKBest    
kbest=selectkbest(indep_X,dep_Y,7)
# created a empty list to append the final value
acclog=[]
accsvml=[]
accsvmnl=[]
accknn=[]
accnav=[]
accdes=[]
accrf=[]
```

```python
kbest # since k=6 inputted, hence got 6 o/p
```

```python
# from the 6 feature we are split X_train, X_test, y_train, y_test & calling split_scalar and i/p 6 feature and dep_y
X_train, X_test, y_train, y_test=split_scalar(kbest,dep_Y) 
    
#running ML algo  and find out classifier,Accuracy,report,X_test,y_test,cm in algo func input x, y and x_test        
classifier,Accuracy,report,X_test,y_test,cm=logistic(X_train,y_train,X_test)
acclog.append(Accuracy) # we can either append classifier or Accuracy or report or X_test or y_test or cm to get score.

classifier,Accuracy,report,X_test,y_test,cm=svm_linear(X_train,y_train,X_test)  
accsvml.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=svm_NL(X_train,y_train,X_test)  
accsvmnl.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=knn(X_train,y_train,X_test)  
accknn.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=Navie(X_train,y_train,X_test)  
accnav.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=Decision(X_train,y_train,X_test)  
accdes.append(Accuracy)
    
classifier,Accuracy,report,X_test,y_test,cm=random(X_train,y_train,X_test)  
accrf.append(Accuracy)
# finally calling selectk_Classification and input neccessary args   
result=selectk_Classification(acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf)
```
